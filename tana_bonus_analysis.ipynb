# Bonus Question 1

Create a list of all transactions from January 2024 sorted by sale amount from highest to lowest.
Identify transactions with sales greater than $500.

```python
# Import pandas for data analysis
import pandas as pd

# Import matplotlib for charts
import matplotlib.pyplot as plt
```

```python
# Load store sales dataset
store_sales = pd.read_csv("data/StoreSales..csv")

# Convert transaction date column into datetime format
store_sales["Transaction Date"] = pd.to_datetime(
    store_sales["Transaction Date"]
)

# Load store detail dataset
store_detail = pd.read_csv("data/StoreDetail.csv")

# Load products dataset
products = pd.read_csv("data/Products.csv")

# Load customer list dataset
customer_list = pd.read_csv("data/customer_list.csv")
```

```python
# Filter January 2024 transactions

jan_2024_sales = store_sales[
    (store_sales["Transaction Date"].dt.year == 2024) &
    (store_sales["Transaction Date"].dt.month == 1)
]

# Sort by sale amount from highest to lowest
jan_2024_sorted = jan_2024_sales.sort_values(
    by="Sale Amount",
    ascending=False
)

# Display transactions greater than $500
high_sales = jan_2024_sorted[
    jan_2024_sorted["Sale Amount"] > 500
]

high_sales.head(10)
```

This analysis filters all January 2024 transactions and identifies high-value purchases above $500.

Sorting the transactions from highest to lowest helps identify the largest sales transactions during the month.

# Bonus Question 6

Calculate total revenue for each product category across all territories and identify which category generated the highest revenue.

```python
# Merge sales and product data
category_sales = pd.merge(
    store_sales,
    products,
    on="Prod Num"
)

# Calculate total revenue by category
category_revenue = category_sales.groupby("CategoryID")["Sale Amount"].sum()

# Sort highest to lowest
category_revenue = category_revenue.sort_values(ascending=False)

# Display results
category_revenue
```

This analysis calculates total revenue generated by each product category.

The results help identify which categories contribute most to overall business revenue.

# Bonus Question 9

Calculate total revenue by region and identify which region generates the highest revenue.

```python
# Merge store sales with store details
regional_sales = pd.merge(
    store_sales,
    store_detail,
    on="Store ID"
)

# Calculate revenue by region
region_revenue = regional_sales.groupby("Region")["Sale Amount"].sum()

# Sort highest to lowest
region_revenue = region_revenue.sort_values(ascending=False)

# Display results
region_revenue
```

This analysis compares sales revenue across business regions.

Grouping by region helps identify which geographic area contributes the most revenue.
