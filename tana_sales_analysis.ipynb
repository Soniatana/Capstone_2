```python
# Convert transaction date column into datetime format
store_sales["Transaction Date"] = pd.to_datetime(store_sales["Transaction Date"])
```

##This step extracts the month and year from each transaction date to support monthly revenue analysis..

```python
# Create a new month column from transaction dates
store_sales["Month"] = store_sales["Transaction Date"].dt.to_period("M")
```

# STEP 3 — MERGE SALES + STORE DATA

This merge combines transaction-level sales data with store information using the Store ID column.

This allows sales transactions to be connected with:
- states
- store locations
- territory managers
- regions

```python
# Merge sales dataset with store detail dataset

sales_merged = pd.merge(store_sales, store_detail, on="Store ID")
```

# Question 3

# Rank the sales performance of each store in the assigned territories and identify the top-performing stores.

This analysis groups Florida sales data by store location and calculates the total revenue generated by each store.

The `.sort_values(ascending=False)` method ranks stores from highest to lowest sales performance.

# Question 4

# Identify the top customers in each assigned sales territory using Rewards ID and customer information.

```python
# Reload customer list dataset using pipe separator
customer_list = pd.read_csv(
    "data/customer_list.csv",
    sep="|"
)

# Display corrected customer dataset information
customer_list.info()
```

This merge combines customer information with sales transaction data using the RewardsID column.

This allows customer purchases to be connected to customer names and account information.

```python
# Merge sales data with customer information
customer_sales = pd.merge(
    store_sales,
    customer_list,
    left_on="RewardsID",
    right_on="cust_id"
)

# Preview merged customer sales data
customer_sales.head()
```

This merge combines sales transaction data with customer information.

The sales dataset uses `RewardsID` while the customer dataset uses `cust_id`, so `left_on` and `right_on` are used to connect the matching columns.

```python
# Merge customer sales with store detail information
customer_sales_merged = pd.merge(
    customer_sales,
    store_detail,
    on="Store ID"
)

# Preview merged dataset
customer_sales_merged.head()
```

This step combines customer purchase activity with store information such as store location, territory manager, and region.

```python
# Filter Florida customer sales
fl_customer_sales = customer_sales_merged[
    customer_sales_merged["State"] == "Florida"
]

# Calculate top Florida customers
top_fl_customers = fl_customer_sales.groupby("name")["Sale Amount"].sum().sort_values(ascending=False)

# Display top 10 Florida customers
top_fl_customers.head(10)
```

This analysis identifies the customers who generated the highest sales revenue in Florida.

Customers are ranked from highest to lowest total purchase value.

```python
# Filter South Carolina customer sales
sc_customer_sales = customer_sales_merged[
    customer_sales_merged["State"] == "South Carolina"
]

# Calculate top South Carolina customers
top_sc_customers = sc_customer_sales.groupby("name")["Sale Amount"].sum().sort_values(ascending=False)

# Display top 10 South Carolina customers
top_sc_customers.head(10)
```

This analysis identifies the highest-spending customers in South Carolina based on total sales revenue generated.

# Question 5

# Analyze the number of transactions and total sales revenue per month by product category in each assigned sales territory.

```python
# Merge sales data with product information
sales_products = pd.merge(
    store_sales,
    products,
    on="Prod Num"
)

# Preview merged dataset
sales_products.head()
```

This merge connects transaction sales data with product information using the product number column.

This allows product names and categories to be included in the sales analysis.

```python
# Merge product sales with store details
sales_products_merged = pd.merge(
    sales_products,
    store_detail,
    on="Store ID"
)

# Preview merged dataset
sales_products_merged.head()
```

This step combines product sales information with store territory details such as state, region, and store location.

```python
# Calculate Florida transactions by product
fl_product_transactions = sales_products_merged[
    sales_products_merged["State"] == "Florida"
]

fl_product_summary = fl_product_transactions.groupby("Product")["Sale Amount"].agg(
    ["count", "sum"]
).sort_values(by="sum", ascending=False)

# Display top 10 Florida products
fl_product_summary.head(10)
```

This analysis shows the most purchased and highest revenue-generating products in Florida.

The count column represents the number of transactions, while the sum column represents total sales revenue.

```python
# Calculate South Carolina transactions by product
sc_product_transactions = sales_products_merged[
    sales_products_merged["State"] == "South Carolina"
]

sc_product_summary = sc_product_transactions.groupby("Product")["Sale Amount"].agg(
    ["count", "sum"]
).sort_values(by="sum", ascending=False)

# Display top 10 South Carolina products
sc_product_summary.head(10)
```

This analysis identifies the top-performing products in South Carolina based on transaction volume and total sales revenue.
